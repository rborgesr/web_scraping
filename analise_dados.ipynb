```python

# Vamos substituir "Nonefound" por NaN
df.loc[df["producers"].str.contains("Nonefound", na=False), "producers"] = np.nan
```

```python
unique_producers = []
```

```python
# Extrai os producers únicos
for data in df["producers"]:
    if type(data) != float:
        new_data = data.split(";")
        for producer in new_data:
            if producer not in unique_producers:
                unique_producers.append(producer)
```

```python
#Listando só os 25 primeiros
unique_producers[:25]
```

```python
#Cria um dicionario
prod_dic = dict()
```

```python
#Inicia o diciário com zero
for i in unique_producers:
    prod_dic[i] = 0
```

```python
#Agrega os produtores
for prod in unique_producers:
    for i in range(0, len(df["producers"])):
        if type(df["producers"][i]) != float and (df["producers"][i].__contains__(prod)) == True:
            prod_dic[prod] += 1
```

```python
#Total de animes para cada produtor (listando só os 20 primeiros)
# Exibir os primeiros 20 itens do dicionário
list(prod_dic.items())[:20]
```

```python
# Ordenando o resultado
{k: v for k, v in sorted(prod_dic.items(), key = lambda item: item[1])}
```

```python
#Ordenando corretamente (do maior para o menor)
{k: v for k, v in sorted(prod_dic.items(), key=lambda item: item[1], reverse=True)}
```
